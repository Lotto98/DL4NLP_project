```python
import json
import matplotlib.pyplot as plt

with open('output/metrics.json', 'r') as f:
    metrics = []
    for line in f:
        metrics.append(json.loads(line))
        
#with open('output/metrics_val.json', 'r') as f:
#    metrics_val = []
#    for line in f:
#        metrics_val.append(json.loads(line))
```

```python
metrics[0].keys()
```

```python
total_loss = [metric.get('total_loss', None) for metric in metrics]
iterations = [metric.get('iteration', None) for metric in metrics]

plt.plot(iterations[::19], total_loss[::19])
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Train Loss every 20 Iteration')
plt.show()

plt.plot(iterations, total_loss)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Train Loss every Iteration')
plt.show()
```

```python
#sum total loss every tot iterations
epoch_size = 2093
loss_per_epoch = [sum(total_loss[i:i+epoch_size]) for i in range(0, len(total_loss), epoch_size if epoch_size > 0 else 1)]
epochs = [i for i in range(0, len(loss_per_epoch))]

print(loss_per_epoch)

plt.plot(epochs, loss_per_epoch)
plt.xlabel('Epoch')
plt.ylabel('Epoch Loss')
plt.title('Train Loss per Epoch')
plt.show()
```
