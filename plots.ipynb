```python
import json
import matplotlib.pyplot as plt

with open('output_old/metrics.json', 'r') as f:
    metrics = []
    for line in f:
        metrics.append(json.loads(line))
```

```python
metrics[0].keys()
```

```python
total_loss = [metric.get('total_loss', None) for metric in metrics]
iterations = [metric.get('iteration', None) for metric in metrics]

plt.plot(iterations, total_loss)
plt.xlabel('Iterations')
plt.ylabel('Total Loss')
plt.title('Total Loss over Iterations')

#add line at loss = 40
plt.axhline(y=40, color='r', linestyle='--')

#add line at iteration 1000
plt.axvline(x=1000, color='r', linestyle='--')

plt.show()
```

```python
#plot lr over iterations
lr = [metric.get('lr', None) for metric in metrics]
plt.plot(iterations[50:], lr[50:])

#add line at iteration 1000
plt.axvline(x=1000, color='r', linestyle='--')
```
